### ggplot ###
ggplot is an R package for making nice plots. ggplots tend to look nicer than plots made by the default R plotting functions. Here is an example taken from [ggplot examples](http://r-statistics.co/Top50-Ggplot2-Visualizations-MasterList-R-Code.html)

```r
library(ggplot2)
theme_set(theme_classic())
options(repr.plot.width = 6, repr.plot.height = 4)

## the mpg data contains miles per gallon for many car models / years
mpg

```

```r
## long tables are annoying to display, so we usually just print the first few rows
nrow(mpg)
ncol(mpg)
head(mpg)
```

```r
# Plot
g <- ggplot(mpg, aes(cty))
g + geom_density(aes(fill=factor(cyl)), alpha=0.8) + 
    labs(title="Density plot", 
         subtitle="City Mileage Grouped by Number of cylinders",
         caption="Source: mpg",
         x="City Mileage",
         fill="# Cylinders")
```
