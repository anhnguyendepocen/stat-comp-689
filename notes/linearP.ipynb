# Basics of Linear Regression

Important take away points:

* Matrix form of regression equation
* Linear regression can model non--linear relationships by means of basis expansions e.g. polynomials or splines

In simple linear regression we have $n$ observation pairs $(x_i,y_i)$. We write $y_i$ as a linear (or affine) function of $x_i$ with mean 0, uncorrelated errors $\epsilon_i$.
$$ y_i = \beta_1 + \beta_2 x_i + \epsilon_i$$

```python
## generate data from linear model
import numpy as np

n = 100
beta1 = 1.
beta2 = 2.
x = np.random.uniform(low=-2.0,high=2.0,size=100)
ep = np.random.uniform(low=-1.,high=1.0,size=100) ## often errors are assumed normal, but don't have to be
y = beta1 + beta2*x + ep
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')

plt.plot(x, y, 'o', color='black');
```

More generally we can write
$$ Y = X\beta + \epsilon$$
where $Y$ is an $n \times 1$ matrix, $X$ is an $n \times p$ matrix, $\beta$ is a $p \times 1$ matrix, and $\epsilon$ is a $n \times 1$ matrix.

```python
## write in vector form
X = np.matrix((np.ones(n),x)).T
B = np.matrix([beta1,beta2]).reshape((2,1))
E = np.matrix(ep).reshape((n,1))
Y = np.matmul(X,B) + E
```

```python
## change Y into a vector for checking that it matches earlier y
Y = np.array(Y).reshape((100))
np.array_equal(Y,y)
```

We can model non-linear relationships by including functions of the $x$ in the matrix $X$. Once on has more than 1 or 2 predictors, the matrix form of the equations is more compact.

```python
## write in vector form
X = np.matrix((np.ones(n),x,np.power(x,3))).T
B = np.matrix([1.0,2.0,4.0]).reshape((3,1))
E = np.matrix(ep).reshape((n,1))
Y = np.matmul(X,B) + E
```

```python
Y = np.array(Y).reshape((100))
```

```python
plt.plot(x, Y, 'o', color='black');
```

Even though Y is a non-linear function of x is is a linear function of the columns of X, i.e. a linear function of $1$, $x$, $x^2$.
